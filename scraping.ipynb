```python
#use the new env to run the jupyter notebook

import pandas as pd
import numpy as np 
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from lxml import html
import time 
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

Base_url= 'https://it.soccerway.com/'
```

```python
def get_players(link,session):
    page = session.get(Base_url+link, headers=headers)
    #xpath = /html/body/d5)iv[9]/div[2]/div[2]/div/div/div[2]/div[8]/div[1]/table
    print(page.content)
    table= html.fromstring(page.content).xpath('/html/body/div[9]/div[2]/div[2]/div/div/div[2]/div[8]/div[1]/table')
    print(table)

    # Close the brow
```

```python

url="/national/italy/serie-a/2022202/regular-season/r77308/"

with requests.Session() as session:
    response = session.get(Base_url+url, headers=headers)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the table with the class "matches"
        matches_table = soup.find('table', class_='matches')

        # Check if the table is found
        if matches_table:
            # Iterate over rows to extract data
            for row in matches_table.find_all('tr'):
                score_time_cell = row.find('td', class_='score-time')
                if score_time_cell and row.find('td', class_='day').text.strip() =="FT":
                    score_element = score_time_cell.find('a')
                    time_element = score_time_cell.find('span')
                    home_team = row.find('td', class_='team-a').text.strip()
                    away_team = row.find('td', class_='team-b').text.strip()
                    score = score_element.text.strip() if score_element else 'N/A'

                    link = score_element['href'] if score_element else 'N/A'


                    players=get_players(link,session)
                    print(f'{home_team} {score} {away_team} {time_element.text.strip()}')
                    print(f'Link: {link}')
                    print(f'Players: {players}')
                    print('-----------------------------')

        else:
            print("Couldn't find a table with the class 'matches'.")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
